```python
x= data[['Sepallength', 'Sepalwidth', 'Petallength', 'Petalwidth']]
y= data['Species']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.05, random_state= 1)
print('Shape of train set is : ', x_train.shape)
print('Shape of test set is : ', x_test.shape)
```

```python
# Using Decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
# Creating a Decision Tree :
from sklearn import tree
dtc = DecisionTreeClassifier()
dtree = dtc.fit(x_train, y_train)
plt.figure( figsize = (20, 10))
print(tree.plot_tree( dtree ))
```

```python
# Importing necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
# Visualizing the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names= ['Sepallength', 'Sepalwidth', 'Petallength', 'Petalwidth'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
