```python
# Importing libraries
import requests
from bs4 import BeautifulSoup
import pandas
import matplotlib.pyplot as plt
import seaborn
import numpy
%matplotlib inline
```

```python
# URL for the page : Finance Rate on Personal Loans at Commercial Banks, 24 Month Loan
url= 'https://fred.stlouisfed.org/series/TERMCBPER24NS'
```

```python
# Checking page response
response= requests.get(url)
print('Response of the URL is :', response)
```

```python
# Viewing HTML content of the page
html_content= response.content
html_content
```

```python
# Parsing the HTML content
soup= BeautifulSoup(html_content, 'html.parser')
print(soup.prettify)
```

```python
# URL for downloading dataset as csv
url2= 'https://fred.stlouisfed.org/graph/fredgraph.csv?bgcolor=%23e1e9f0&chart_type=line&drp=0&fo=open%20sans&graph_bgcolor=%23ffffff&height=450&mode=fred&recession_bars=on&txtcolor=%23444444&ts=12&tts=12&width=1168&nt=0&thu=0&trc=0&show_legend=yes&show_axis_titles=yes&show_tooltip=yes&id=TERMCBPER24NS&scale=left&cosd=1972-02-01&coed=2021-08-01&line_color=%234572a7&link_values=false&line_style=solid&mark_type=diamond&mw=3&lw=2&ost=-99999&oet=99999&mma=0&fml=a&fq=Monthly&fam=avg&fgst=lin&fgsnd=2020-02-01&line_index=1&transformation=lin&vintage_date=2021-10-15&revision_date=2021-10-15&nd=1972-02-01'
```

```python
# Checking response of the URL 2
# ETL : Extract, Transform, Load
response= requests.get(url2)
print('Response of the second URL is :', response)
html_content= response.content
print(html_content)
```

```python
# Parsing the HTML content
soup= BeautifulSoup(html_content, 'html.parser')
print(soup.prettify)
```

```python
# Appending soup elements into a list for analysis
data= []
for x in soup:
  data.append(x)
print(data)
```

```python
# Replacing '\n' from the list
data = [x.strip().split('\n') for x in data]
print(data)
print('Length of the list is :', len(data))
```

```python
# Conversion of the list from double list to single list for further analysis
data2= []
for sublist in data:
  for x in sublist:
    data2.append(x)
print(data2)
print('Length of the list is :', len(data2))
```

```python
# Creating a dataframe 'finance_data' out of the list data2
finance_data = pandas.DataFrame({'Values': data2})
# Viewing first 2 rows of the dataframe
finance_data.head(2)
```

```python
# Shape of the dataframe
finance_data.shape
```

```python
# Dropping value at first index number as it is an unwanted value
finance_data.drop(index= finance_data.index[0], axis= 0, inplace= True)
finance_data.head(2)
```

```python
# Creating 2 columns 'Date' and 'Rate' out of the 'Values' column
finance_data[['Date','Rate']] = finance_data['Values'].str.split(',', expand= True)
del finance_data['Values']
```

```python
# Data types of the dataframe
finance_data.dtypes
```

```python
# Converting 'Date' to datatime datatype
finance_data['Date'] = pandas.to_datetime(finance_data['Date'] )
# Converting 'Rate' to numeric datatype
finance_data['Rate'] = pandas.to_numeric(finance_data['Rate'], errors= 'coerce' )
# Checking data types again
finance_data.dtypes
```

```python
# Checking null values in the dataframe
finance_data.isnull().sum()
```

```python
# Dropping null values rows
finance_data.dropna(inplace= True)
```

```python
# Checking null values again in the dataframe
finance_data.isnull().sum()
```

```python
# Checking duplicate values in the dataframe
finance_data.duplicated().sum()
```

```python
# After dropping rows, some index numbers are lost. So resetting the index numbers
finance_data= finance_data.reset_index()
# Deleting unwanted 'index' column which is generated after resetting the index numbers
del finance_data['index']
```

```python
# Viewing first 10 rows of our final dataframe : time series data
finance_data.head(10)
```

```python
# Checking shape of the dataframe again
finance_data.shape
```

```python
# Brief information about dataframe
finance_data.info()
```

```python
# Ignoring warnings
import warnings
warnings.filterwarnings('ignore')
# Plotting Scatterplot between date and rate % variables
plt.figure(figsize= (20, 7))
seaborn.scatterplot(finance_data['Date'], finance_data['Rate'], color= 'blue')
plt.title('Scatterplot between Date and Rate %', fontsize= 14, fontweight= 'bold')
plt.xlabel('Date', fontsize= 13, fontweight= 'bold')
plt.ylabel('Rate %', fontsize= 13, fontweight= 'bold')
plt.xticks(fontsize= 12, fontweight= 'bold')
plt.yticks(fontsize= 12, fontweight= 'bold')
plt.show()
```

```python
# Plotting Lineplot between date and rate % variables
plt.figure(figsize= (20, 7))
plt.plot(finance_data['Date'], finance_data['Rate'], color= 'red')
plt.title('Lineplot between Date and Rate %', fontsize= 14, fontweight= 'bold')
plt.xlabel('Date', fontsize= 13, fontweight= 'bold')
plt.ylabel('Rate %', fontsize= 13, fontweight= 'bold')
plt.xticks(fontsize= 12, fontweight= 'bold')
plt.yticks(fontsize= 12, fontweight= 'bold')
plt.show()
```

```python
# Description of the dataframe
finance_data.describe()
```

```python
# Plotting histogram plot for rate % variable
axes= finance_data.hist(figsize= (20, 7), color= 'green')
plt.title('Histogram plot for Rate %', fontsize= 14, fontweight= 'bold')
plt.xlabel('Values for Rate %', fontsize= 13, fontweight= 'bold')
plt.ylabel('Frequency of occurence', fontsize= 13, fontweight= 'bold')
plt.xticks(fontsize= 12, fontweight= 'bold')
plt.yticks(fontsize= 12, fontweight= 'bold')
plt.show()
```

```python
# Distribution plot for Rate % variable in the dataframe. 
# kde (Kernel Density Estimate) : It gives us probability density function
# bins denotes in how many parts our data is distributed
plt.figure(figsize= (20,5))
seaborn.distplot(finance_data['Rate'], bins= 10, kde=True)
plt.show()
```

```python
# Control chart of Mean Rate percentage vs Observed Rate percentage
plt.figure(figsize= (20, 5) )
plt.axhline(y = finance_data['Rate'].mean(), color = 'green', linestyle = '-', linewidth= 1.5, label= 'Mean Rate %')
plt.axhline(y = finance_data['Rate'].min(), color = 'red', linestyle = '--', linewidth= 1.5)
plt.axhline(y = finance_data['Rate'].max(), color = 'red', linestyle = '--', linewidth= 1.5)
finance_data['Rate'].sort_values(ascending= True).plot(color= 'blue', label= 'Observed Rate %')
plt.title('Control chart between Mean Rate percentage and Observed Rate percentage', fontsize= 14, fontweight= 'bold')
plt.xlabel('Index Number of observations', fontsize= 12, fontweight= 'bold')
plt.ylabel('Observed Rate percentage', fontsize= 12, fontweight= 'bold')
plt.xticks(fontsize= 12, fontweight= 'bold')
plt.yticks(fontsize= 12, fontweight= 'bold')
plt.legend()
plt.show()
```

```python
# Violin plot of Rate % variable
plt.figure(figsize= (10, 5) )
plt.violinplot(finance_data['Rate'], showmedians= True)
plt.show()
```

```python
# Boxplot for the variable
plt.figure(figsize= (10, 5))
plt.boxplot(finance_data['Rate'])
plt.xlabel('Boxplot', fontsize= 12, fontweight= 'bold')
plt.ylabel('Rate values', fontsize= 12, fontweight= 'bold')
plt.yticks(fontsize= 12, fontweight= 'bold')
plt.grid()
plt.show()
```
