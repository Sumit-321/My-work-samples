```python
# Correlation in our dataset
correlation= data.corr().round(decimals= 3)
correlation
# I am seeing a correlation of 0.479 between variables Sales and Profit. 
```

```python
# Plotting heatmap
plt.figure(figsize=(23, 5))
seaborn.heatmap(correlation, cmap='RdYlGn', annot= True)
plt.show()
```

```python
# Scatterplot
plt.figure(figsize= (20,5))
seaborn.scatterplot(x= data['Profit'], y=data['Sales'], color= 'black')
plt.show()
```

```python
# Relplot
seaborn.relplot(x= data['Profit'], y=data['Discount'], color= 'orange')
plt.show()
```

```python
# Regression plot
plt.figure(figsize= (15,5))
seaborn.regplot(x= data['Profit'], y=data['Sales'], marker= '.', color= 'green')
plt.show()
```

```python
# Countplot
plt.figure(figsize= (20, 5))
seaborn.countplot(x= data['Quantity'], hue= data['Segment'], palette= ['red', 'yellow', 'green'])
plt.show()
```
